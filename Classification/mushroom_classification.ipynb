## Importing the Libraries

```python
import numpy as np
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('./../Data/mushrooms.csv')

dataset.describe()
```

```python
dataset.info()
```

## Encoding categorical data

```python
dataset = pd.get_dummies(dataset).astype('int')
```

```python
X = dataset.iloc[:, 1:].values
y = dataset.iloc[:, 1].values
```

## Spliting the dataset into Training set & Test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Training the dataset with Logistic regression

```python
from sklearn.linear_model import LogisticRegression

regressor_li = LogisticRegression()
regressor_li.fit(X_train,y_train)
```

## Using cross validation

```python
from sklearn.model_selection import cross_val_score

cross_val_score(regressor_li, X, y, cv=4).mean()
```

## Predict test set results

```python
y_pred_li = regressor_li.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred_li)
```

## Using Random Forest Algorithm

```python
from sklearn.ensemble import RandomForestClassifier

regressor_rf = RandomForestClassifier()
cross_val_score(regressor_rf, X, y, cv=4).mean()
```
